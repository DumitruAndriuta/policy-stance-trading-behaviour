```python
# Step 1: Read Proximity Data from Text File
file_path = 'investmentproximity.txt'  # Replace with the path to your text file
with open(file_path, 'r') as f:
    lines = f.readlines()
    
# Step 2: Create Proximity Matrix
proximity_matrix = []
for line in lines:
    proximity_matrix.append([float(val) for val in line.split()])

# Step 3: Convert Proximity Matrix to Graph
G = nx.from_numpy_array(np.array(proximity_matrix))

# Step 4: Apply Fruchterman-Reingold Layout
pos = nx.spring_layout(G)

# Step 5: Visualize the Network
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10)
plt.title('Voting Proximity Network')
plt.show()
```

```python
# Step 1: Read Proximity and Party Affiliation Data from Text File
proximity_file = 'investmentproximity.txt'  # Replace with the path to your proximity data file
party_file = 'rounded_partyaffiliation.txt'   # Replace with the path to your party affiliation data file

with open(proximity_file, 'r') as f:
    proximity_lines = f.readlines()

party_affiliation = {}    
with open(party_file, 'r') as f:
    party_lines = f.readlines()

# Step 2: Create Proximity Matrix
proximity_matrix = []
for line in proximity_lines:
    numbers = [float(val) for val in line.split()]
    proximity_matrix.append(numbers)

# Convert proximity_matrix to numpy array
proximity_matrix = np.array(proximity_matrix)



party_affiliation = {}

# Assuming each line contains multiple integers representing party affiliation for each politician
for i, line in enumerate(party_lines):
    # Split the line into individual integers
    party_list = [int(val) for val in line.strip().split()]
    for j, party in enumerate(party_list):
        party_affiliation[i*len(party_list) + j] = 'Republican' if party == 1 else 'Democrat'

# Step 3: Convert Proximity Matrix to Graph and Add Node Attributes
G = nx.from_numpy_array(np.array(proximity_matrix))

# Add party affiliation as node attributes
nx.set_node_attributes(G, party_affiliation, 'party')

# Step 4: Apply Fruchterman-Reingold Layout
pos = nx.spring_layout(G)

# Step 5: Visualize the Network with Color-Coded Nodes for Party Affiliation
plt.figure(figsize=(8, 6))

# Separate nodes by party affiliation for coloring
republican_nodes = [node for node, data in G.nodes(data=True) if data['party'] == 'Republican']
democrat_nodes = [node for node, data in G.nodes(data=True) if data['party'] == 'Democrat']

# Draw Republican nodes
nx.draw_networkx_nodes(G, pos, nodelist=republican_nodes, node_color='red', node_size=500, alpha=0.8)

# Draw Democrat nodes
nx.draw_networkx_nodes(G, pos, nodelist=democrat_nodes, node_color='blue', node_size=500, alpha=0.8)

plt.axis('off')
plt.show()
```
